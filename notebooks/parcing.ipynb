### Импорт библиотек

```python
import requests
from bs4 import BeautifulSoup as BS
```

### Парсим данные 

```python
r = requests.get("https://www.marktplaats.nl/q/lego+poppetjes/")
html = BS(r.content, 'html.parser')
```

### Проверяем получены ли данные 

```python
if html != 0:
    kek = "Данные получены"
    
print(kek)
```

### Выводим названия, стоимость всех полученных объектов

```python
page = 1

while True:
    r = requests.get("https://www.marktplaats.nl/q/lego+poppetjes/p/" + str(page) + "/")
    html = BS(r.content, 'html.parser')
    pages = html.select(".hz-Listings--list-view > .hz-Listing--list-item ")
    
    if (len(pages)):
        for el in pages:
            tittle = el.find("h3", {"class": "hz-Listing-title"}).text
            cost = el.find("span", {"class": "hz-Listing-price"}).text
            print(tittle, cost)
        page += 1
        print(page)
    else:
        break
```

```python
for el in html.select(".hz-Listings--list-view > .hz-Listing--list-item "):
    img = el.find("figure", {"class": "hz-Listing-image-container"})
    print(img)
```
